```python
import pandas as pd
import re
```

```python
Data_frame ='/full.csv'
df = pd.read_csv (Data_frame)
df['Name'] = df['Name'].str.upper()
df
```

```python
def extrair_primeiro_nome(nome):
    if ". " in nome:
      primeiro_nome = nome.split(". ")[1].split(" ")[0]
      return primeiro_nome
    else:
        return ''
df['Primeiro_nome'] = df['Name'].apply(extrair_primeiro_nome)
```

```python
nome_frequencia = df.groupby('Class')['Primeiro_nome'].agg(lambda x: x.value_counts().idxmax())
contagem_frequencia = df.groupby('Class')['Primeiro_nome'].agg(lambda x: x.value_counts().max())


resultado = pd.DataFrame({
    'Classe': nome_frequencia.index,
    'Primeiro_nome_mais_comum': nome_frequencia.values,
    'Frequencia': contagem_frequencia.values
})


resultado
```
